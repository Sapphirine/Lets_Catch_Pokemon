## Plot longitude and latitude

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
from mpl_toolkits.basemap import Basemap
```

```python
true_latitude = pd.read_csv("../Result/true_latitude.csv")
true_longitude = pd.read_csv("../Result/true_longitude.csv")
# data from Yi Qin
predict_latitude = pd.read_csv("../Result/predict_latitude.csv")
predict_longitude = pd.read_csv("../Result/predict_longitude.csv")
# data from Ge Qu
predict_latitude_q = pd.read_csv("../Result/latitude_result.csv")
predict_longitude_q = pd.read_csv("../Result/longitude_result.csv")
```

```python
plt.figure(figsize=(20,10))
m1 = Basemap(projection='merc',
             llcrnrlat=-60,
             urcrnrlat=65,
             llcrnrlon=-180,
             urcrnrlon=180,
             lat_ts=0,
             resolution='c')


m1.fillcontinents(color='#303030',lake_color='#000000') # dark grey land, black lakes
m1.drawmapboundary(fill_color='#000000')                # black background
m1.drawcountries(linewidth=0.2, color="w")              # thin white line for country borders

# Plot the data
# x,y true
x,y = m1(true_longitude['True_longitude'].tolist(),true_latitude['True_latitude'].tolist())
# x_1, y_1 from Yi Qin
x_1, y_1 = m1(predict_longitude['true longitude'].tolist(),predict_latitude['Predict_latitude'].tolist())
# x_2, y_2 from Ge Qu
x_2,y_2 = m1(predict_longitude_q['0'].tolist(),predict_latitude_q['0'].tolist())
m1.scatter(x_2,y_2, s=3, c="#1292db", lw=0, alpha=1, zorder=5)
m1.scatter(x,y, s=3, c='r', lw=0,alpha=1,zorder=5)
plt.title("Pokemon activity")
plt.savefig("../pic/Pokemon activity noclass")
plt.show()
```

## Plot city

```python
true = pd.read_csv("../Result/city.csv")
true.head()
# len(true)
predict = pd.read_csv("../Result/Predict_city.csv")
predict.head()
# len(predict)
```

```python
true = pd.read_csv("../Result/city_label.csv")
predict = pd.read_csv("../Result/city_pred.csv")
print(len(true))
```

```python
cnt = true.groupby('0')['0'].size()
cnt_p = predict.groupby('0')['0'].size()
# cnt['label']='0'
# print(cnt)
# print(cnt_p.head())
```

```python
# cnt = cnt.to_frame()
cnt_index = np.expand_dims(np.array(cnt.index),axis=1)
cnt_values = np.expand_dims(np.array(cnt.values),axis=1)
print ("cnt_index:len{}\n{}".format(len(cnt_index),cnt_index))
print ("cnt_values:len()\n{}".format(len(cnt_index),cnt_values))
label = np.full((len(cnt_index),1),'true')
cnt_true = np.concatenate((cnt_index,cnt_values,label),axis=1)
# cnt = np.array(cnt)
# cnt.reindex(columns={cnt.columns[1]:'value'},inplace=True)
print(cnt_true)
```

```python
# cnt = cnt.to_frame()
cnt_p_index = np.expand_dims(np.array(cnt.index),axis=1)
cnt_p_values = np.expand_dims(np.array(cnt.values),axis=1)
print ("cnt_index:len{}\n{}".format(len(cnt_index),cnt_index))
print ("cnt_values:len()\n{}".format(len(cnt_index),cnt_values))
label_p = np.full((len(cnt_index),1),'predict')
cnt_predict = np.concatenate((cnt_p_index,cnt_p_values,label_p),axis=1)
# cnt = np.array(cnt)
# cnt.reindex(columns={cnt.columns[1]:'value'},inplace=True)
print(cnt_predict)
```

```python
cnt_all = np.concatenate((cnt_true,cnt_predict),axis=0)
cnt_all_df = pd.DataFrame(cnt_all,columns=['city','values','label'])
cnt_all_df[['values']]= cnt_all_df[['values']].apply(pd.to_numeric)
cnt_all_df
# print(type(cnt_all_df['values']))
```

## Plot Pokemon

```python
Names = pd.read_csv("../Data/pokemonNumbers.csv",header=None)
Names.columns = ['pokemonid', 'name']
Names['pokemonid'] = Names['pokemonid'].apply(lambda x: x-1)
Names.head()
dictionary = dict(zip(Names['pokemonid'], Names['name']))
```

```python
idlabel = pd.read_csv("../Result/id_label.csv")
idpredict = pd.read_csv("../Result/id_pred.csv")
```

```python
id_l_cnt = idlabel.groupby('true')['true'].size()
```

```python
idpredict.head()
id_l_cnt.head()
id_l_cnt.index
```

```python
id_p_cnt = idpredict.groupby('predict')['predict'].size()
id_p_cnt.head()
id_p_cnt.index
```

```python
id_all = pd.concat([id_l_cnt,id_p_cnt],axis=1)
id_all.fillna(0,inplace=True)
# id_all.rename(columns={id_all.columns[0]:"index",id_all.columns[1]:"true",id_all.columns[0]:"predict"})
id_all.head()
```

```python
data = (
    id_all.set_index(id_all.index)
      .stack()  # un-pivots the data 
      .reset_index()  # moves all data out of the index
      .rename(columns={'level_0':'pokemon','level_1': 'Variable', 0: 'Value'})
)
data[['Value']] = data[['Value']].apply(pd.to_numeric)
data
```

```python
pokemon_id = data.pokemon
pokemon_id
pokemon_name = []
for i in pokemon_id:
    pokemon_name.append(dictionary[i])
```

```python
f, ax = plt.subplots(figsize=(20, 10))
sns.barplot(x=pokemon_name, y=data['Value'], hue=data['Variable'], data=data, ax=ax)
plt.xticks(rotation=90)
plt.setp(ax.patches, linewidth=0)
plt.savefig('../pic/id_predict')
texts = ax.set(title="number of id")
```

```python
f, ax = plt.subplots(figsize=(10, 30))
sns.barplot(y=pokemon_name, x=data['Value'], hue=data['Variable'], data=data,orient='h',ax=ax)
plt.xticks(rotation=90)
plt.setp(ax.patches, linewidth=0)
plt.savefig('../pic/id_predict_h')
texts = ax.set(title="number of id")
```
